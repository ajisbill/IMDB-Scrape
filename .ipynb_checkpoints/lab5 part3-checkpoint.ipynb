```python
import bs4
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

### Merging Dataframes
* Here, we read the csv files created in parts 1 and 2 into their own dataframes. 
* Then, we use `df.merge()` which merges the data frames on the column "Titles".
* This creates one dataframe with all movies that are contained within both dataframes (given they have the same title).

```python
the_numbers = pd.read_csv("the_numbers.csv", encoding='latin-1')
imdb = pd.read_csv("imdb.csv", encoding='latin-1')
```

```python
df = imdb.merge(the_numbers, on="Titles")
df
```

### Cleaning the Data
* We use `df.apply` with the function `pd.to_numeric` to convert all columns with numerical data to floats or ints.
    * **Note**: Some cleaning steps were performed during parsing in parts 1 and 2.
* We convert the "Release Date" column to datetime to be used in time series analysis.
* "Gross" column is dropped because it represents the same data as "TN Domestic Gross" but is less precise.

```python
cols = df.columns.drop(['Titles', 'Genre', 'Release Date'])
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df["Release Date"] = pd.to_datetime(df["Release Date"])
df = df.drop(columns ="Gross")
df
```

## Analysis
**Steps:**
* `df.describe()` gives a general overview of the data so we can look at stats such as average, max, and min of all columns.
* Using a scatter matrix on the numerical data gives us a general idea of which columns are corellated.
    * We can then use a scatter plot with two specific columns to look at a particular correlation.


```python
df.describe()
```

```python
pd.plotting.scatter_matrix(df[["User Rating","Worldwide Gross", "Budget", "Votes", "TN Domestic Gross", "Metascore"]], figsize= (14,14));
```

### Scatter Matrix Analysis
* The matrix yields expected results such as the strong correlation Domestic Gross and Worldwide Gross and between User Rating and Metascore. 
    * This confirms that our data is valid, as we would expect to large correlation coefficients between these related columns.
* Utilizing data from both IMDB and The Numbers we can look at specific correlations such as Worldwide Gross vs. Budget and User Rating vs. Worldwide Gross.

### Worldwide Gross vs. Budget

```python
df.describe()
r = df["Budget"].corr(df["Worldwide Gross"]) 

sns.regplot(x = df["Budget"],y= df["Worldwide Gross"], scatter_kws={'s':8}).set(title="Worldwide Gross vs. Budget" + "  \nR=" +str(r));
plt.rcParams["figure.figsize"] = (8,6)
```

### Analysis
* Plotting Worldwide Gross vs. Budget yields a medium correlation (r=0.59).
* We see that **as budget increases, worldwide gross generally increases** as well.
    * However, worldwide gross does not always increase with increases in budget.
    * There are some vertical lines which indicates movies with the same budget that had vastly different worldwide gross figures.
* There are also notable outliers such as movies that had a low budget and returned an above average worldwide gross, along with high budget movies with small gross numbers (movies that "flopped").

### User Rating vs. Worldwide Gross

```python
r = df["Worldwide Gross"].corr(df["User Rating"])
sns.regplot(x = df["Worldwide Gross"],y= df["User Rating"], scatter_kws={'s':8}).set(title="User Rating vs. Worldwide Gross" + "  \nR=" +str(r));
plt.rcParams["figure.figsize"] = (8,6)
```

### Analysis
* We see an **even weaker correlation between User Rating and Worldwide gross** (r=0.37).
* The general trend indicates that movies with higher gross figures tend to have higher user ratings.
    * However, there is a cluster of data around the $25-100 million mark of movies with similar gross revenues but with various user ratings (3-8.5).
    * This tells us that how much money a movie made does not necessarily correlate with how users rated the movie.

### Plotting Movie Budgets Over Time
* Using the "Release Date" column, we use `.resample('Y')` to look at the average yearly movie budgets from 1996 to 2020.
* `.relplot()` is used to create a continuous line plot to see the change in movie budgets over time.

```python
budget_year = df.set_index('Release Date')
budget_year = budget_year.resample('Y').mean()
```

```python
sns.relplot(x="Release Date", y="Budget", kind="line", data=budget_year).set(title="Average Movie Budget Over Time");
```

### Analysis
* As expected, average movie budgets have increased approximately linearly over the last 25 years.
* From 1996 to 2020, **average movie budgets have increased by $81 million**.
* Further analysis, might include adjusting for inflation or comparing the increases in movie budget over different genres.
